# 1. Зниження розмірності і візуалізація даних
Застосуйте методи зниження розмірності sklearn.decomposition.PCA і sklearn.manifold.TSNE для візуалізації даних, з якими ви працювали в лабораторній № 1 (знижуючи розмірність до двох). Візуалізуйте результат.

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

data = pd.read_csv('heart.csv')

# Попередня обробка даних (кодування категоріальних змінних)
categorical_features = ['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope']
data_encoded = pd.get_dummies(data, columns=categorical_features, drop_first=True)

# Відокремлення ознак та нормалізація
X = data_encoded.drop(columns=['HeartDisease'])
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

# PCA для зниження розмірності до двох компонентів
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_normalized)

# TSNE для зниження розмірності до двох компонентів
tsne = TSNE(n_components=2, random_state=42, init='pca', learning_rate='auto')
X_tsne = tsne.fit_transform(X_normalized)

# Візуалізація PCA
plt.figure(figsize=(10, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=data_encoded['HeartDisease'], cmap='viridis', s=10)
plt.title('PCA Visualization')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.colorbar(label='HeartDisease')
plt.show()

# Візуалізація TSNE
plt.figure(figsize=(10, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=data_encoded['HeartDisease'], cmap='viridis', s=10)
plt.title('TSNE Visualization')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.colorbar(label='HeartDisease')
plt.show()
```

PCA намагається максимізувати варіацію даних у кожному вимірі, і ми можемо побачити, що групи даних із різними значеннями HeartDisease (0 і 1) частково перетинаються, але також мають певний ступінь розділення. Це показує те, що ознаки у датасеті містять певну інформацію, яка може бути корисною для класифікації.

Друга візуалізація показує більш чітке розділення між групами. TSNE фокусується на збереженні локальних структур даних, тому ми бачимо, що точки, що належать до однієї групи, згруповані в компактні кластери. Підхід дозволяє краще виявляти приховані структури у даних, але може бути менш інтерпретованим.

# 2. Кластерний аналіз
За допомогою алгоритму k-means зробіть квантування зображення (видалення візуально надлишкової інформації) з глибиною 64, 32, 16 та 8 рівнів для будь-якого обраного самостійно зображення.

```python
from sklearn.cluster import KMeans
import numpy as np
from PIL import Image

image = Image.open('dog.png')
image_np = np.array(image)

# Функція для квантування зображення
def quantize_image(image, n_colors):
    # Перетворення зображення в формат (ширина * висота, 3)
    pixels = image.reshape(-1, 3)
    
    # k-means
    kmeans = KMeans(n_clusters=n_colors, random_state=42)
    labels = kmeans.fit_predict(pixels)
    
    # Відновлення зображення на основі кластерів
    quantized_pixels = kmeans.cluster_centers_[labels].astype('uint8')
    quantized_image = quantized_pixels.reshape(image.shape)
    
    return quantized_image

# Кквантування з різною кількістю рівнів
depths = [64, 32, 16, 8]
quantized_images = {depth: quantize_image(image_np, depth) for depth in depths}

# Візуалізація
plt.figure(figsize=(15, 10))
plt.subplot(2, 3, 1)
plt.imshow(image)
plt.title("Original")
plt.axis('off')

for i, (depth, quantized_image) in enumerate(quantized_images.items(), start=2):
    plt.subplot(2, 3, i)
    plt.imshow(quantized_image)
    plt.title(f"{depth} levels")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

Оригінальне зображення зберігає всі деталі та плавні переходи кольорів, а зменшення кількості кольорів поступово знижує деталізацію. При 64 рівнях зображення майже не втрачає якості, при 32 і 16 рівнях спрощуються кольорові області, а при 8 зображення стає значно менш деталізованим, але залишається впізнаваним.

# 3. Обробка та класифікація текстових даних
Завантажте набір текстових даних (з мітками класів). Проведіть передобробку даних (видаліть стоп-слова, пунктуацію), за допомогою wordcloud зробіть візуалізацію найбільш поширених слів або n-gram у кожному класі. Векторизуйте тексти (наприклад за допомогою sklearn.feature_extraction.text.TfidfVectorizer). Проведіть класифікацію текстових даних, зробіть оцінку якості. 

```python
import pandas as pd
import string
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import re

data = pd.read_csv('ecommerceDataset.csv')
data.head()
```

```python
# Перейменування колонок для зручності
data.columns = ['Category', 'Description']

# Видалення HTML-розмітки, пунктуації та очищення тексту
def preprocess_text(text):
    text = re.sub(r'<.*?>', '', text)  # Видалення HTML
    text = text.lower()  # Перетворення в нижній регістр
    text = text.translate(str.maketrans('', '', string.punctuation))  # Видалення пунктуації
    return text

# Очищення текстових даних
data['Cleaned_Description'] = data['Description'].astype(str).apply(preprocess_text)

# Побудова WordCloud для найбільш частих слів у описах
all_texts = ' '.join(data['Cleaned_Description'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_texts)
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title("WordCloud для описів товарів")
plt.axis('off')
plt.show()

# Перевірка категорій у колонці 'Category'
print(data['Category'].value_counts())
```

Найчастіші слова, такі як "book", "use", "make", "home", "product", відображають загальний характер описів.

```python
# Векторизація текстів за допомогою TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=1000)
X = vectorizer.fit_transform(data['Cleaned_Description']).toarray()
y = data['Category']  # Категорії товарів як цільова змінна

# Розділення на навчальну і тестову вибірки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Навчання моделі класифікації (Random forest)
classifier = RandomForestClassifier(random_state=42)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)
print(classification_report(y_test, y_pred))
print("Confusion matrix:")
print(confusion_matrix(y_test, y_pred))
```

Ми бачимо, що опис товарів у категорії Books часто включає слова, пов’язані з літературою та читанням, наприклад, "book", "read", "story". У категорії "Electronics" домінують слова такі як "device", "use", "battery". Категорія "Household" містить слова, що описують товари для дому, наприклад, "wall", "decor", "painting", у "Clothing & Accessories" зустрічаються слова на кшталт "style", "fabric", "design".

Загальна точність моделі становить 96%, а F1-score для всіх категорій знаходиться в межах 0.96-0.97. Матриця плутанини показує, що модель правильно класифікувала більшість описів, хоча невелика кількість помилок мається між схожими категоріями, такими як "Household" і "Electronics".
